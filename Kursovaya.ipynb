# Критерий Фрацини (модельные данные) 



```python
def froc(x):
    n = len(x)
    x_mean = np.mean(x)
    s = 0
    for i in range(n):
        s+=(x[i]-x_mean)**2
    s = s/n

    b_n = 0
    for i in range(n):
        z = (x[i]-x_mean)/math.sqrt(s)
        b_n += abs(norm.cdf(z)-((i-0.5)/n))
    b_n = b_n/math.sqrt(n)
    return b_n
```

# это равномероное распределение, оно нам не надо

```python
import numpy as np
import random
N = 1000
n1 = 10
n2 = 100
n3 = 5000

pirs = []
for i in range(100):
    arr = [random.randint(0, 100) for i in range(1000)]
    k = len(arr)
    m = 100 # количество выборок
    h = (max(arr) - min(arr))/(m-1)
    b = np.arange(min(arr)+h, max(arr)+h, (max(arr) - min(arr))/(m-1))

    x = (arr-b.reshape(-1,1))
    x = ((x<0)|(x==0)).sum(axis=1)
    x = np.hstack(([x[0]], np.diff(x)))
    b -= h/2

    x_ = (x*b).sum()/k

    s = np.sqrt(((b - x_)**2 * x).sum()/(k-1))

    u = (b - x_)/s
    phi = np.exp(-u*u/2)/np.sqrt(2*np.pi)*k*h/s

    hi = ((x - phi)**2/phi).sum()
    pirs.append(round(hi, 3))
    # Критическая точка распределения Χ^2
    # при len(x) - 3 == 7 равна ~ 14.1
    # данные подчиняются нормальному закону распределения
#print(pirs)
import matplotlib.pyplot as plt

plt.hist(pirs)
```

#Другим популярным графиком для проверки распределения выборки данных является график квантиль-квантиль, график Q-Q или график QQ для краткости.

Этот график генерирует собственную выборку идеализированного распределения, с которым мы сравниваем, в данном случае распределение Гаусса. Идеализированные образцы делятся на группы (например, 5), называемые квантилями. Каждая точка данных в выборке связана с аналогичным элементом из идеализированного распределения с тем же кумулятивным распределением.
